```python
from langchain_community.chat_models import ChatOllama
```

```python
import json
```

```python
model = ChatOllama(model="llama3")
```

```python
essay = """
Artificial Intelligence is transforming modern society. It helps automate tasks,
improve healthcare diagnostics, and enhance decision-making in many industries.
However, ethical concerns such as bias and job displacement remain important issues.
"""
```

```python
prompt = f"""
Evaluate the following essay.

Return ONLY JSON in this format:
Do not include explanations.
Do not include text before or after JSON.
{{
 "score": integer from 1 to 10,
 "feedback": string
}}

Essay:
{essay}
"""
```

```python
response = model.invoke(prompt)
```

```python
print(response.content)
```

```python
data = json.loads(response.content)
```

```python
print(data['score'])
```
